En este notebook me encargo solamente de obtener las masas efectivas isotrópicas. Hay distitntos casos. según [wikipedia](https://en.wikipedia.org/wiki/Effective_mass_(solid-state_physics))
hay una masa efectiva de la densidad de estados:
$$
m^{*}_{density} = \sqrt[3]{g^2 m_x m_y m_z}
$$

donde $g$ es un factor de degeneración que dice la multiplicidad de valles en la zona de brillouin.

o tambien está la masa efectiva segun el modelo de Drude:
$$
m^{*}_{conductivity} = \frac{1}{3}\left[ 
\frac{1}{m_x} +
\frac{1}{m_y} +
\frac{1}{m_z} +
\right]^{-1}
$$
pero, esta última vale er raras ocaciones porque la ley de drude depende tambien de la movilidad de los portadores.

## masa del electron

```python
m_e = databands[(databands['mass']=='m_e')]
m_e_gx_up = m_e[(m_e['toname']=='(X)') & (m_e['spin']=='(up)')]
m_e_gx_dn = m_e[(m_e['toname']=='(X)') & (m_e['spin']=='(down)')]
m_e_gz_up = m_e[(m_e['toname']=='(Z)') & (m_e['spin']=='(up)')]
m_e_gz_dn = m_e[(m_e['toname']=='(Z)') & (m_e['spin']=='(down)')]
```

```python
# guardo las tablas a txts
with open('m_e.dat', 'w') as me:
    me.write('\n=====gamma->X up\n')
m_e_gx_up.to_csv('m_e.dat', sep=' ', index=False, header=True,mode='a')
with open('m_e.dat', 'a') as me:
    me.write('\n=====gamma->X down\n')
m_e_gx_dn.to_csv('m_e.dat', sep=' ', index=False, header=True,mode='a')
with open('m_e.dat', 'a') as me:
    me.write('\n=====gamma-Z Up\n')
m_e_gz_up.to_csv('m_e.dat', sep=' ', index=False, header=True,mode='a')
with open('m_e.dat', 'a') as me:
    me.write('\n=====gamma-Z Down\n')
m_e_gz_dn.to_csv('m_e.dat', sep=' ', index=False, header=True,mode='a')
```
